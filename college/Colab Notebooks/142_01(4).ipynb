```
# Import Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
from torch.autograd import Variable
from torch.utils.data import DataLoader
```

```
import cv2
import os

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```

```
features_train=load_images_from_folder('/content/fruit/training/apple')
targets_train=[]
for l in range(len(features_train)):
  targets_train.append('apple')
print(len(targets_train))
for l in range(len(features_train)):
  targets_train.append('orange')
print(len(targets_train))
a=load_images_from_folder('/content/fruit/training/orange')
for x in a:
  features_train.append(x)
print(len(features_train))


features_test=load_images_from_folder('/content/fruit/test/apple')
targets_test=[]
for l in range(len(features_test)):
  targets_test.append('apple')
print(len(targets_test))
for l in range(len(features_test)):
  targets_test.append('orange')
print(len(targets_test))
a=load_images_from_folder('/content/fruit/test/orange')
for x in a:
  features_test.append(x)
print(len(features_test))
```

```
features_train, features_test = np.array(features_train, np.float32), np.array(features_test, np.float32)
features_train=features_train.reshape(198,100*100*3)
print(features_train.shape)
features_test=features_test.reshape(66,100*100*3)
print(features_test.shape)
```

```
features_train, features_test = features_train / 255., features_test / 255
```

```
import tensorflow as tf
featuresTrain = tf.convert_to_tensor(features_train, dtype=tf.float32)
featuresTest = tf.convert_to_tensor(features_test, dtype=tf.float32)

targetsTrain = tf.convert_to_tensor(targets_train)
targetsTest = tf.convert_to_tensor(targets_test)
```

```
# batch_size, epoch and iteration
batch_size = 100
n_iters = 5000
num_epochs = n_iters / (len(features_train) / batch_size)
num_epochs = int(num_epochs)

# Pytorch train and test sets
train = torch.utils.data.TensorDataset(featuresTrain,targetsTrain)
test = torch.utils.data.TensorDataset(featuresTest,targetsTest)

# data loader
train_loader = DataLoader(train, batch_size = batch_size, shuffle = False)
test_loader = DataLoader(test, batch_size = batch_size, shuffle = False)

# visualize one of the images in data set
plt.imshow(featuresTrain[10].reshape(100,100,3))
plt.axis("off")
plt.title(str(targetsTrain[10]))
plt.savefig('graph.png')
plt.show()
```
